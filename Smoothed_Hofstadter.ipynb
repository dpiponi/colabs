```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=np.load('hex.data.npy')
```

```python
m.shape
```

```python
plt.imshow(m)
```

```python
m[:,0]
```

```python
e=np.linspace(-4, 4, 3025)[:,None]*np.ones(1024)
e.shape
```

```python
img = np.zeros((3025, 1024))
for i in range(0, 3025, 1):
  if i % 100 == 0:
    print(i)
  img = img + 1/(0.00001 + (e - m[:, i])**2)
plt.figure(figsize=(24, 16))
plt.imshow(img, aspect = 0.5, cmap='plasma')
```

```python
plt.figure(figsize=(24, 16))
plt.imshow(img, aspect = 0.25, cmap='inferno',vmin=0,vmax=3000000)
```
